<h3><b>(T.9) Factorials</b></h3>
<h3><b>Note: Starting from this task it is mandatory that all remaining problems are completed using functions.</b></h3>
<p>
Create a factorial function that returns the factorial of a numbers. The function should accept only positive integers. </br>
The factorial of a number is the sum of the multiplication, of all the whole numbers, from our specified number down to 1. </br>
Example: </br>
The factorial of 6 would be 6 x 5 x 4 x 3 x 2 x 1 = 720

</p>


```python
def factorial(num):
    sum = 1
    for x in reversed(range(1,num+1)):
        sum = sum*x
    return sum

 
print(factorial(6))
```

<h3><b>(T.14) Pyramid</b></h3>
<p>
Print the following pattern on the screen where n = 5: </br>


        *
       ***
      *****
     *******
    *********
</p>


```python
def pyramid(n):
    stars=1
    spaces=n-1
    for i in range(0,n):
    
        print(' ' * spaces + "*" * stars)
        stars=stars+2
        spaces=spaces-1

pyramid(24)
```

<h3><b>(T.15) Inverted pyramid</b></h3>
<p>
Print the following pattern on the screen where n = 5: </br>
    
    *********
     *******
      *****
       ***
        *

</p>


```python
def pyramid_rev(n):
    spaces=0
    stars=n*2-1
    for i in range(0,n):
    
        print(' ' * spaces + "*" * stars)
        stars=stars-2
        spaces=spaces+1

pyramid_rev(7)
```

answer: longest

```python

def longest_word(sentence):
    split_words=sentence.split()
    curr=0
    longest=""

    for word in split_words:
        count_words=len(word)

        if curr>count_words:
            curr=curr
            
        else: 
            curr=count_words
            longest=word
        
    print(longest)


longest_word("This is the longest word")
```

<h3><b>(T.23) Decimal to Hexadecimal</b></h3>

Example: </br>
decimal = 123 </br>
hexadecimal = 7B</br>

Note: DO NOT USE BUILT-IN Python functions!!! Think about using the operators. Check the hex numerical system.

</p>


```python
def decimal_to_hexadec(num):
    hex_digits = "0123456789ABCDEF"
    result = ""

    if num == 0:
        return "0"
    while num > 0:
        remainder = num % 16
        result = hex_digits[remainder] + result
        num = num // 16
    return result   

decimal_to_hexadec(12314)
```

is_prime returns True

```python
def is_prime(num):
    if num < 2:
        return False
        
    for i in range(2, int(num**0.5)+ 1 ):
        if num % i == 0:
            return False

    else:
        return True
    

is_prime(2)

```

<h3><b>(T.27) Sort an array of 0s, 1s and 2s</b></h3>
<p>

Create a function that sorts 0s, 1s and 2s in order</br>

Example: </br>
Input: [0, 1, 2, 0, 1, 2] </br>
Output: [0, 0, 1, 1, 2, 2] </br>
Explanation: [0, 0, 1, 1, 2, 2] has all 0s first, then all 1s and all 2s in last.



</p>


```python

import array as arr

def sort_array(my_arr):
    x = arr.array('i', my_arr)

    a = x.count(0)
    b = x.count(1)
    c = x.count(2)

    sorted_array=('['+'0,' * a + '1,' * b + '2,' * c)
    sorted_array=sorted_array.rstrip(',') + ']'
    print(sorted_array)

sort_array([0,1,2,0,1,2])

```
